```python
syn = IP(dst='www.google.com')/TCP(dport=80, flags='S')
```

```python
syn_ack = sr1(syn)
```

```python
getStr = 'GET / HTTP/1.1\r\nHost: www.google.com\r\n\r\n'
```

```python
print(syn_ack[TCP].dport)
```

```python
req2 = IP(dst='www.google.com') / TCP(dport=80, sport=syn_ack[TCP].dport, seq=syn_ack[TCP].ack, ack=syn_ack[TCP].seq + 1, flags='A') / getStr
reply = sr1(req2)
```
